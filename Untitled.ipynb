```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

DATA ANALYSIS OF CUSTOMER SHPPING DATA

This information will enable an individual interested in opening a business in this area know which business to set up.

We will answer questions like
1. What item is mostly bought?
2. Which mall is most popular?
3. Which gender shops the most


```python
df = pd.read_csv('customer_shopping_data.csv')
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()

df.category.unique()
```

VISUALIZATION

```python
plt.figure(figsize=(12, 9))
plt.hist(df.category)
plt.title('MOST POPULAR ITEM SHOPPED');
```

```python
plt.figure(figsize=(15, 9))
plt.hist ( df.shopping_mall )
plt.title('MOST POPULAR SHOPPING MALL')
```

```python
plt.figure(figsize=(12, 9))
sns.barplot(x='category', y='price', data=df)
plt.title('MOST EXPENSIVE ITEMS BOUGHT');
```

```python
sns.pairplot(df, hue= 'gender')
plt.tight_layout(pad=2);
```

FROM DATA ANALYSIS, WE HAVE DISCOVERED THAT

1. What item is mostly bought? Clothes
2. Which mall is most popular? Kanyon and Mall of Istanbull
3. Which gender spends the most money when shopping? Males

Therefore an individual may want to venture into clothing business and locate it at Kanyon mall or Istanbull mall. The focus could be on male when selecting high-end items for sell
